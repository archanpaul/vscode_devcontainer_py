```python
"""
Test script to check whether all required packages from pyproject.toml are installed.
"""

import importlib

# Packages from pyproject.toml
packages = [
    # Core
    "requests",
    # Web group
    "fastapi",
    "flask",
    "jinja2",
    "sqlmodel",
    "uvicorn",
    # ML group
    "adjustText",
    "ipykernel",
    "ipywidgets",
    "jupyter",
    "keras",
    "matplotlib",
    "numpy",
    "pandas",
    "torch",
    "sklearn",
    "seaborn",
    "streamlit",
    "tensorflow",
    "torchvision",
    # Dev group
    "pytest",
    "black",
]


def check_packages(pkgs):
    results = {}
    for pkg in pkgs:
        try:
            importlib.import_module(pkg)
            results[pkg] = "✓ Installed"
        except ImportError as e:
            results[pkg] = f"✗ Missing ({e})"
    return results


def run_package_tests():
    results = check_packages(packages)
    print("\nPackage installation check:\n")
    for pkg, status in results.items():
        print(f"{pkg:15} : {status}")


run_package_tests()
```
